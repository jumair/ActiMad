```python
# Hago las importaciones necesarias
import requests
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Creo una variable "ruta" con la ruta donde dejaré los archivos 
ruta = 'D:/GitHub/ActiMad/'
```

```python
# Pido el archivo de las actividades culturales de Madrid por API
resp = requests.post('https://datos.madrid.es/egob/catalogo/206974-0-agenda-eventos-culturales-100.csv')
print(resp.status_code)
```

```python
# Y lo creo en la misma carpeta con el nombre de actividades-100.csv
with open( ruta + 'actividades-100.csv', 'w' ) as f:
            f.write( resp.text )
            f.close()
```

```python
# Lo abro con pandas y muestro las primeras 5 filas
results = pd.read_csv(ruta + "actividades-100.csv", delimiter=";",encoding = "ISO-8859-1")
results.head(5)
```

```python
# saco el nombre de las columnas y veo que la primera columna tiene un espacio en blanco antes del nombre : ID-EVENTO
results.columns
```

```python
# Me quedo con todas las filas pero sólo de 10 columnas y las ordeno por el BARRIO-INSTALACION
acti = results.loc[:, [" ID-EVENTO", "TITULO", "GRATUITO", "HORA", "DISTRITO-INSTALACION", "BARRIO-INSTALACION", "CODIGO-POSTAL-INSTALACION", "LONGITUD", "LATITUD", "TIPO"]].sort_values(by="BARRIO-INSTALACION")
acti.head(5)
```

```python
titulo = "TITULO"
gratuito = "GRATUITO"
hora = "HORA"
distrito = "DISTRITO-INSTALACION"
barrio = "BARRIO-INSTALACION"
codigo = "CODIGO-POSTAL-INSTALACION"
longitud = "LONGITUD"
latitud = "LATITUD"
tipo = "TIPO"

# Renombramos 10 columnas de 2f ormas diferentes y ponemos inplace para que lo cambie sobre el mismo acti y no nos de una copia
acti.rename(columns={' ID-EVENTO': "Evento"}, inplace=True)
acti.rename(columns={titulo: "Titulo", gratuito: "Gratuito", hora: "Hora", distrito: "Distrito", barrio: "Barrio", codigo: "Codigo"}, inplace=True)
acti.rename(columns={longitud: "Longitud", latitud: "Latitud", tipo: "Tipo"}, inplace=True)
acti.head(5)
```

```python
#Número filas y columnas
acti.shape
```

```python
# Borrar todas las filas que contengan NaN y lo dejo en dataframe actividades, pero nosotros sólo vamos a quitar las filas con NaN en 2 columnas con subset
actividades = acti.dropna(subset=["Latitud", "Longitud"])
actividades.shape
```

```python
actividades.info()
```

```python
# Limpiar el código para quitarle el .0 que es un float64, cambiarlo a texto como distrito
actividades.loc[:,'Codigo'] = actividades['Codigo'].astype(str)
```

```python
actividades.info()
```

```python
actividades.head(5)
```

```python
# Limpiar el código para quitarle el .0 mejor con función lambda como texto cogiendo 5 primeros caracteres porque lo anterior no funciona.
actividades['Codigo'] = actividades['Codigo'].apply(lambda x : x[:5])
actividades.head(5)
```

```python
# Saca las actividades unicas de la columna Tipo
actividades.Tipo.unique()
```

```python
# Dejo en acti la copia de actividades, borro los nan de la columna Tipo y le aplico una función lambda para quitarle el primer caracter '/' a la 
# columna Tipo
acti = actividades
acti = acti.dropna(subset=["Tipo"])
acti['Tipo'] = acti['Tipo'].apply(lambda x : x[1:])
acti.head(5)
```

```python
# Voy a crear un dataframe tipos de la columna Tipo separando su contenido por el carácter '/' y creará 4 columnas en el dataframe tipos
tipos = acti['Tipo'].str.split('/', expand = True)
# Renombro las columnas creadas
tipos.columns = ['Tipoa', 'Tipob', 'Tipo1', 'Tipo2']
# Borro las columnas 1 y 2 porque tienen los valores contenido y actividades respectivamente y no aportan nada. 
# Me quedo sólo con las columnas Tipo1 y Tipo2.
tipos.drop(["Tipoa", "Tipob"], axis = 1, inplace = True)

# Al dataframe acti le concateno el nuevo dataframe creado llamado "tipos"
acti = pd.concat([acti, tipos], axis=1)
acti.head(5)
```

```python
# Aquí tengo el dataframe preparado con todas las filas que tienen longitud y latitud para poder trabajar con el y mostrarlo en un mapa
# Guardar el resultado en el fichero actividades-100-fin.csv en la misma carpeta usando la variable ruta, los campos están separados por comas
# Poner al final index=False para que no incluya el numero índice de cada fila
actividades_fin = acti.to_csv(ruta + 'actividades-100-fin.csv', index=False)

# En este fichero tengo las actividades culturales de Madrid que tienen latitud y longitud para poder representarlas en un mapa. 
```
